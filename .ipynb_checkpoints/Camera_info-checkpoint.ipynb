```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
import pickle

from IPython.display import display
from ipywidgets import interactive
# from ipywidgets import IntSlider

def camera_calibration():
    # 1. Load individual images from a specified folder
    image_folder = "D:/spkim/coding/python/yolov8_sp/cali"
    image_paths = [os.path.join(image_folder, file) for file in os.listdir(image_folder) if file.endswith(".jpg")]
    
    camera_images = [cv2.imread(path) for path in image_paths]
    
    # Ensure all images are loaded correctly
    for i, img in enumerate(camera_images):
        if img is None:
            raise FileNotFoundError(f"Image at {image_paths[i]} could not be loaded.")
    
    # Define ROIs for each camera in the composite image
    camera_rois = [
        ( 0, 0, 1200, 1000), # Camera 1 ROI (x, y, width, height)
        ( 1200, 0, 1200, 1000), # Camera 2 ROI
        ( 2400, 0, 1200, 1200), # Camera 3 ROI
        ( 0, 1000, 1200, 1000), # Camera 4 ROI
        ( 1200, 1000, 1200, 1000) # Camera 5 ROI
    ]
    
    # Extract individual camera images using the ROIs
    camera_images = [
        img[y:y+h, x:x+w]
        for img, (x, y, w, h) in zip(camera_images, camera_rois)
    ]
    
    # 2. Camera Calibration
    # Define calibration parameters (example values, use actual calibration data)
    object_points = []  # 3D points in real-world space
    image_points = []   # 2D points in image plane
    
    # Prepare object points for a chessboard pattern (example: 9x6 grid)
    chessboard_size = (6, 9)
    square_size = 1.0  # Size of a square in your defined unit (e.g., meters)
    objp = np.zeros((chessboard_size[0] * chessboard_size[1], 3), np.float32)
    objp[:, :2] = np.mgrid[0:chessboard_size[0], 0:chessboard_size[1]].T.reshape(-1, 2)
    objp *= square_size
    
    for img in camera_images:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        ret, corners = cv2.findChessboardCorners(gray, chessboard_size, None)
    
        if ret:
            # Refine corner locations for sub-pixel accuracy
            corners = cv2.cornerSubPix(
                gray, corners, (11, 11), (-1, -1),
                criteria=(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
            )
            object_points.append(objp)
            image_points.append(corners)
    
            # Optional: Draw and display the corners for debugging
            # cv2.drawChessboardCorners(img, chessboard_size, corners, ret)
            # cv2.imshow('Chessboard Corners', img)
            # cv2.waitKey(500)
        else:
            print("Chessboard not detected in one of the images.")
    
    # cv2.destroyAllWindows()
    
    # Perform camera calibration for each camera
    calibration_results = []
    for i, (img, img_points) in enumerate(zip(camera_images, image_points)):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(
            [object_points[i]], [img_points], gray.shape[::-1], None, None
        )
        calibration_results.append({
            "mtx": mtx,
            "dist": dist,
            "rvecs": rvecs,
            "tvecs": tvecs
        })
    
    # Save calibration results for future use with pickle
    import pickle
    with open("camera_calibration_results.pkl", "wb") as f:
        pickle.dump(calibration_results, f)
    
    # 3. Use calibration data for 3D reconstruction
    # Example: Compute disparity and reproject to 3D using the first two cameras
    img_left = camera_images[0]
    img_right = camera_images[1]
    
    mtx_left = calibration_results[0]["mtx"]
    dist_left = calibration_results[0]["dist"]
    mtx_right = calibration_results[1]["mtx"]
    dist_right = calibration_results[1]["dist"]
    
    # Ensure both images are the same size
    height, width = img_left.shape[:2]
    img_right = cv2.resize(img_right, (width, height))
    
    # Rectify the images if necessary
    # (rectification logic depends on stereo calibration, not included here)

# 필터링된 2D 점과 해당하는 카메라 보정 결과를 기반으로 삼각측량
def triangulate_points(points_2d, calibration_results):
    """
    Triangulate 3D points using 2D points from multiple cameras, considering visibility.

    Parameters:
    points_2d (list): List of (x, y) coordinates from visible cameras. Use None for missing points.
    calibration_results (list): Camera calibration results containing intrinsic and extrinsic parameters.

    Returns:
    np.ndarray: The triangulated 3D point.
    """
    valid_points = []
    valid_projection_matrices = []

    # 필터링: 보이는 카메라만 사용
    for point, calib in zip(points_2d, calibration_results):
        if point is not None:  # 해당 카메라에서 점이 보이는 경우만 추가
            valid_points.append(point)
            mtx = calib["mtx"]
            rvec = calib["rvecs"][0]
            tvec = calib["tvecs"][0]

            # Compute rotation matrix from rotation vector
            rot_matrix, _ = cv2.Rodrigues(rvec)

            # Construct projection matrix: P = K [R|t]
            projection_matrix = np.dot(mtx, np.hstack((rot_matrix, tvec)))
            valid_projection_matrices.append(projection_matrix)

    if len(valid_points) < 2:
        raise ValueError("At least two cameras must see the point to triangulate.")

    # Prepare 2D points for triangulation
    points_2d_homogeneous = np.array([[x, y, 1] for x, y in valid_points]).T

    # Perform triangulation using the first two valid cameras
    points_3d_homogeneous = cv2.triangulatePoints(
        valid_projection_matrices[0], valid_projection_matrices[1],
        points_2d_homogeneous[:2, 0], points_2d_homogeneous[:2, 1]
    )

    # Convert from homogeneous coordinates to 3D coordinates
    points_3d = points_3d_homogeneous[:3] / points_3d_homogeneous[3]

    return points_3d.flatten()

# Data pre-processing: 0 or None value is replaced by nearest value
def data_pre_processing(points_data):
    # Treat 0 as missing (convert 0 to np.nan)
    points_data[points_data == 0] = np.nan
    # print(points_data)
    # Forward fill
    for i in range(1, len(points_data)):
        mask = np.isnan(points_data[i, :])
        points_data[i, mask] = points_data[i - 1, mask]
    
    # Backward fill
    for i in range(len(points_data) - 2, -1, -1):
        mask = np.isnan(points_data[i, :])
        points_data[i, mask] = points_data[i + 1, mask]
    
    # Load calibration results from pickle
    with open("camera_calibration_results.pkl", "rb") as f:
        calibration_results = pickle.load(f)

    return points_data

# Load points from file and process
input_file = "D:/spkim/data/20241114_40Hz_test/20241114_40Hz_test/test.txt"  # File containing points
points_data = np.loadtxt(input_file, delimiter="\t")  # Load the data from file

camera_calibration()

with open("camera_calibration_results.pkl", "rb") as f:
    calibration_results = pickle.load(f)

# Data pre-processing
points_data = data_pre_processing(points_data)

# Extract 2D points and confidences
triangulated_points = []
for row in points_data:
    points_2d = []
    triangulated_row = []  # Store triangulated points for one row
    cnt = 1
    for i in range(0, len(row), 3):
        x, y, conf = row[i:i+3]
        if conf > 0.25:  # Use points with sufficient confidence
            points_2d.append((x, y))
        else:
            points_2d.append(None)  # Mark as None if confidence is low
        if(cnt == 5):
            triangulated_row.extend(triangulate_points(points_2d, calibration_results))
            points_2d = []
            cnt = 0
        cnt += 1
    
    triangulated_points.append(triangulated_row)
    
# Save triangulated points to a text file
output_file = "triangulated_points.txt"
np.savetxt(output_file, triangulated_points, fmt="%.6f", comments="", delimiter="\t")

# # Initialize the figure and axis
# fig = plt.figure(figsize=(8, 6))
# ax = fig.add_subplot(111, projection='3d')
# scatter_plots = []

# # Define the plotting function
# def initialize_plot():
#     global scatter_plots
#     row = triangulated_points[0]  # Start with the first frame
#     colors = plt.cm.tab10(np.arange(len(row) // 3) % 10)  # Use tab10 colormap for distinct colors
#     scatter_plots.clear()
#     for i, color in zip(range(0, len(row), 3), colors):
#         x, y, z = row[i], row[i + 1], row[i + 2]
#         scatter = ax.scatter(x, y, z, c=[color], marker='o')
#         scatter_plots.append(scatter)
#     ax.set_xlabel('X-axis')
#     ax.set_ylabel('Y-axis')
#     ax.set_zlabel('Z-axis')
#     ax.set_title("3D Points - Frame 1")

# # Update the plot dynamically without refreshing the entire figure
# def update_plot(frame_idx):
#     row = triangulated_points[frame_idx]
#     for scatter, i in zip(scatter_plots, range(0, len(row), 3)):
#         x, y, z = row[i], row[i + 1], row[i + 2]
#         scatter._offsets3d = ([x], [y], [z])
#     ax.set_title(f"3D Points - Frame {frame_idx + 1}")
#     plt.draw()

# # Initialize the plot
# initialize_plot()

# # Create an interactive slider and bind the update function
# interactive_plot = interactive(update_plot, frame_idx=(0, len(triangulated_points) - 1))
# display(interactive_plot)

############################################
# Define the plotting function

def plot_frame_with_slider(frame_idx):
    plt.clf()  # Clear the current figure
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    row = triangulated_points[frame_idx]
    colors = plt.cm.tab10(np.arange(len(row) // 3) % 10)  # Use tab10 colormap for distinct colors
    for i, color in zip(range(0, len(row), 3), colors):
        x, y, z = row[i], row[i + 1], row[i + 2]
        ax.scatter(x, y, z, c=[color], marker='o')
    
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title(f"3D Points - Frame {frame_idx + 1}")
    plt.show()

# Create an interactive slider
interactive_plot = interactive(plot_frame_with_slider, frame_idx=(0, len(triangulated_points) - 1))
display(interactive_plot)
```

```python
import cv2
import numpy as np
import os

# 설정
CHESSBOARD_SIZE = (9, 6)  # 체스보드 내부 코너 개수
SQUARE_SIZE = 25  # 각 체스보드 사각형의 한 변 길이 (단위: mm)
calibration_images_path = "./cali"  # 캘리브레이션 이미지 폴더 경로

# 각 카메라의 ROI (x, y, width, height)
camera_rois = [
    (0, 0, 1200, 1000),       # Camera 1 ROI
    (1200, 0, 1200, 1000),    # Camera 2 ROI
    (2400, 0, 1200, 1200),    # Camera 3 ROI
    (0, 1000, 1200, 1000),    # Camera 4 ROI
    (1200, 1000, 1200, 1000)  # Camera 5 ROI
]

# 월드 좌표계에서 체스보드 점 생성
objp = np.zeros((CHESSBOARD_SIZE[0] * CHESSBOARD_SIZE[1], 3), np.float32)
objp[:, :2] = np.mgrid[0:CHESSBOARD_SIZE[0], 0:CHESSBOARD_SIZE[1]].T.reshape(-1, 2)
objp *= SQUARE_SIZE

# 각 카메라별 캘리브레이션 데이터를 저장할 리스트
object_points = []  # 월드 좌표계의 점
image_points = [[] for _ in range(len(camera_rois))]  # 각 카메라의 이미지 좌표

# 이미지 로드 및 처리
image_files = sorted([f for f in os.listdir(calibration_images_path) if f.endswith('.jpg')])

for image_file in image_files:
    image_path = os.path.join(calibration_images_path, image_file)
    image = cv2.imread(image_path)
    if image is None:
        print(f"이미지 {image_file}를 로드할 수 없습니다. 건너뜁니다.")
        continue

    for cam_idx, roi in enumerate(camera_rois):
        x, y, w, h = roi
        cam_image = image[y:y+h, x:x+w]  # ROI로 이미지 분할
        gray = cv2.cvtColor(cam_image, cv2.COLOR_BGR2GRAY)

        # 체스보드 코너 찾기
        ret, corners = cv2.findChessboardCorners(gray, CHESSBOARD_SIZE, None)
        if ret:
            if len(image_points[cam_idx]) == 0:
                object_points.append(objp)  # 첫 번째 이미지에서만 추가

            # 코너 보정
            cv2.cornerSubPix(
                gray, corners, (11, 11), (-1, -1),
                criteria=(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
            )
            image_points[cam_idx].append(corners)

            # 시각화
            cv2.drawChessboardCorners(cam_image, CHESSBOARD_SIZE, corners, ret)
            cv2.imshow(f"Camera {cam_idx + 1}", cam_image)
            cv2.waitKey(200)

cv2.destroyAllWindows()

# 각 카메라 캘리브레이션 수행
camera_matrices = []
dist_coeffs = []
rot_vectors = []
trans_vectors = []

for cam_idx in range(len(camera_rois)):
    if len(image_points[cam_idx]) > 0:
        ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(
            object_points, image_points[cam_idx], gray.shape[::-1], None, None
        )
        camera_matrices.append(mtx)
        dist_coeffs.append(dist)
        rot_vectors.append(rvecs)
        trans_vectors.append(tvecs)
        print(f"Camera {cam_idx + 1}: 캘리브레이션 완료")
    else:
        raise ValueError(f"Camera {cam_idx + 1}의 캘리브레이션에 충분한 데이터가 없습니다.")

# 캘리브레이션 결과 저장 (선택 사항)
np.savez("camera_calibration_data.npz", 
         camera_matrices=camera_matrices, 
         dist_coeffs=dist_coeffs, 
         rot_vectors=rot_vectors, 
         trans_vectors=trans_vectors)

print("모든 카메라의 캘리브레이션 완료. 3D 재구성을 진행할 준비가 되었습니다.")
```

```python
import cv2
import numpy as np
import os

# 설정
CHESSBOARD_SIZE = (9, 6)  # 체스보드 내부 코너 개수
SQUARE_SIZE = 25  # 각 체스보드 사각형의 한 변 길이 (단위: mm)
calibration_images_path = "./cali"  # 캘리브레이션 이미지 폴더 경로

# 각 카메라의 ROI (x, y, width, height)
camera_rois = [
    (0, 0, 1200, 1000),       # Camera 1 ROI
    (1200, 0, 1200, 1000),    # Camera 2 ROI
    (2400, 0, 1200, 1200),    # Camera 3 ROI
    (0, 1000, 1200, 1000),    # Camera 4 ROI
    (1200, 1000, 1200, 1000)  # Camera 5 ROI
]

# 월드 좌표계에서 체스보드 점 생성
objp = np.zeros((CHESSBOARD_SIZE[0] * CHESSBOARD_SIZE[1], 3), np.float32)
objp[:, :2] = np.mgrid[0:CHESSBOARD_SIZE[0], 0:CHESSBOARD_SIZE[1]].T.reshape(-1, 2)
objp *= SQUARE_SIZE

# 각 카메라별 캘리브레이션 데이터를 저장할 리스트
object_points = []  # 월드 좌표계의 점 (이미지당 한 번 추가)
image_points = [[] for _ in range(len(camera_rois))]  # 각 카메라의 이미지 좌표

# 이미지 로드 및 처리
image_files = sorted([f for f in os.listdir(calibration_images_path) if f.endswith('.jpg')])

for image_file in image_files:
    image_path = os.path.join(calibration_images_path, image_file)
    image = cv2.imread(image_path)
    if image is None:
        print(f"이미지 {image_file}를 로드할 수 없습니다. 건너뜁니다.")
        continue

    # 한 이미지에서 체스보드가 감지된 경우만 추가
    object_points_appended = False

    for cam_idx, roi in enumerate(camera_rois):
        x, y, w, h = roi
        cam_image = image[y:y+h, x:x+w]  # ROI로 이미지 분할
        gray = cv2.cvtColor(cam_image, cv2.COLOR_BGR2GRAY)

        # 체스보드 코너 찾기
        ret, corners = cv2.findChessboardCorners(gray, CHESSBOARD_SIZE, None)
        if ret:
            # 한 번만 추가
            if not object_points_appended:
                object_points.append(objp)
                object_points_appended = True

            # 코너 보정
            cv2.cornerSubPix(
                gray, corners, (11, 11), (-1, -1),
                criteria=(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
            )
            image_points[cam_idx].append(corners)

            # 시각화
            cv2.drawChessboardCorners(cam_image, CHESSBOARD_SIZE, corners, ret)
            cv2.imshow(f"Camera {cam_idx + 1}", cam_image)
            cv2.waitKey(200)

cv2.destroyAllWindows()

# 각 카메라 캘리브레이션 수행
camera_matrices = []
dist_coeffs = []
rot_vectors = []
trans_vectors = []

for cam_idx in range(len(camera_rois)):
    if len(image_points[cam_idx]) > 0:
        ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(
            object_points, image_points[cam_idx], gray.shape[::-1], None, None
        )
        camera_matrices.append(mtx)
        dist_coeffs.append(dist)
        rot_vectors.append(rvecs)
        trans_vectors.append(tvecs)
        print(f"Camera {cam_idx + 1}: 캘리브레이션 완료")
    else:
        raise ValueError(f"Camera {cam_idx + 1}의 캘리브레이션에 충분한 데이터가 없습니다.")

# 캘리브레이션 결과 저장 (선택 사항)
np.savez("camera_calibration_data.npz", 
         camera_matrices=camera_matrices, 
         dist_coeffs=dist_coeffs, 
         rot_vectors=rot_vectors, 
         trans_vectors=trans_vectors)

print("모든 카메라의 캘리브레이션 완료. 3D 재구성을 진행할 준비가 되었습니다.")
```

```python
import cv2
import numpy as np
import os
import json

# 설정
CHESSBOARD_SIZE = (9, 6)  # 체스보드 내부 코너 개수
SQUARE_SIZE = 25  # 각 체스보드 사각형의 한 변 길이 (단위: mm)
calibration_images_path = "./cali"  # 캘리브레이션 이미지 폴더 경로

# 각 카메라의 ROI (x, y, width, height)
camera_rois = [
    (0, 0, 1200, 1000),       # Camera 1 ROI
    (1200, 0, 1200, 1000),    # Camera 2 ROI
    (2400, 0, 1200, 1200),    # Camera 3 ROI
    (0, 1000, 1200, 1000),    # Camera 4 ROI
    (1200, 1000, 1200, 1000)  # Camera 5 ROI
]

# 월드 좌표계에서 체스보드 점 생성
objp = np.zeros((CHESSBOARD_SIZE[0] * CHESSBOARD_SIZE[1], 3), np.float32)
objp[:, :2] = np.mgrid[0:CHESSBOARD_SIZE[0], 0:CHESSBOARD_SIZE[1]].T.reshape(-1, 2)
objp *= SQUARE_SIZE

# 각 카메라별 캘리브레이션 데이터를 저장할 리스트
object_points_per_camera = [[] for _ in range(len(camera_rois))]  # 월드 좌표계의 점
image_points = [[] for _ in range(len(camera_rois))]  # 각 카메라의 이미지 좌표

# 이미지 로드 및 처리
image_files = sorted([f for f in os.listdir(calibration_images_path) if f.endswith('.jpg')])

for image_file in image_files:
    image_path = os.path.join(calibration_images_path, image_file)
    image = cv2.imread(image_path)
    if image is None:
        print(f"이미지 {image_file}를 로드할 수 없습니다. 건너뜁니다.")
        continue

    for cam_idx, roi in enumerate(camera_rois):
        x, y, w, h = roi
        cam_image = image[y:y+h, x:x+w]  # ROI로 이미지 분할
        gray = cv2.cvtColor(cam_image, cv2.COLOR_BGR2GRAY)

        # 체스보드 코너 찾기
        ret, corners = cv2.findChessboardCorners(gray, CHESSBOARD_SIZE, None)
        if ret:
            # 월드 좌표계 점 추가 (해당 카메라의 데이터에만 추가)
            if len(object_points_per_camera[cam_idx]) < len(image_points[cam_idx]) + 1:
                object_points_per_camera[cam_idx].append(objp)

            # 코너 보정
            cv2.cornerSubPix(
                gray, corners, (11, 11), (-1, -1),
                criteria=(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
            )
            image_points[cam_idx].append(corners)

            # 시각화
            cv2.drawChessboardCorners(cam_image, CHESSBOARD_SIZE, corners, ret)
            cv2.imshow(f"Camera {cam_idx + 1}", cam_image)
            cv2.waitKey(200)

cv2.destroyAllWindows()

# 각 카메라 캘리브레이션 수행
camera_matrices = []
dist_coeffs = []
rot_vectors = []
trans_vectors = []

for cam_idx in range(len(camera_rois)):
    if len(image_points[cam_idx]) > 0:
        ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(
            object_points_per_camera[cam_idx], image_points[cam_idx], gray.shape[::-1], None, None
        )
        camera_matrices.append(mtx)
        dist_coeffs.append(dist)
        rot_vectors.append(rvecs)
        trans_vectors.append(tvecs)
        print(f"Camera {cam_idx + 1}: 캘리브레이션 완료")
    else:
        print(f"Camera {cam_idx + 1}: 캘리브레이션 데이터를 건너뜁니다.")

# 캘리브레이션 결과 저장
output_data = {
    "camera_matrices": [mtx.tolist() for mtx in camera_matrices],
    "dist_coeffs": [dist.tolist() for dist in dist_coeffs],
    "rot_vectors": [[rvec.tolist() for rvec in rvecs] for rvecs in rot_vectors],
    "trans_vectors": [[tvec.tolist() for tvec in tvecs] for tvecs in trans_vectors],
}

# JSON 파일로 저장
with open("camera_calibration_data.json", "w") as f:
    json.dump(output_data, f, indent=4)

print("모든 사용 가능한 카메라의 캘리브레이션 완료. 결과가 JSON 파일로 저장되었습니다.")

# # 캘리브레이션 결과 저장 (선택 사항)
# np.savez("camera_calibration_data.npz", 
#          camera_matrices=camera_matrices, 
#          dist_coeffs=dist_coeffs, 
#          rot_vectors=rot_vectors, 
#          trans_vectors=trans_vectors)

# print("모든 사용 가능한 카메라의 캘리브레이션 완료. 3D 재구성을 진행할 준비가 되었습니다.")
```
